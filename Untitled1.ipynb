```python
#Preprocessing
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#Creating directories
train_dir="cats_and_dogs_filtered/train"
validation_dir="cats_and_dogs_filtered/validation"
```

```python
#Objects of the class
train_datagen=ImageDataGenerator(rescale=1/255)#0 to 255 pixels
validation_datagen=ImageDataGenerator(rescale=1/255)
```

```python
#Preprocessing
train_generator=train_datagen.flow_from_directory(train_dir,target_size=(150,150))
```

```python
validation_generator=validation_datagen.flow_from_directory(validation_dir,target_size=(150,150))
```

```python
#model building 
model_1=tf.keras.Sequential()
model_1.add(tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)))
model_1.add(tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
model_1.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
model_1.add(tf.keras.layers.Conv2D(filters=128,kernel_size=(3,3),activation='relu'))
model_1.add(tf.keras.layers.Conv2D(filters=128,kernel_size=(3,3),activation='relu'))
model_1.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
model_1.add(tf.keras.layers.Conv2D(filters=256,kernel_size=(3,3),activation='relu'))
model_1.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
model_1.add(tf.keras.layers.Flatten())
model_1.add(tf.keras.layers.Dense(50,activation='relu'))
model_1.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
#Compiling
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=['accuracy'])
```

```python
#GlobalAveragePooling->Flattening,1000 images classification
#Transfer Learning model
```

```python
model_3=tf.keras.Sequential()#top->rightmost,not to take the rightmost part as we are taking only 2 classes not 1000
model_3.add(ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3)))
model_3.add(tf.keras.layers.GlobalMaxPool2D())
model_3.add(tf.keras.layers.Dense(50,activation='relu'))
model_3.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
model_3.summary()
```

```python
model_3=tf.keras.Sequential()#top->rightmost,not to take the rightmost part as we are taking only 2 classes not 1000
model_3.add(ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3)))
model_3.add(tf.keras.layers.GlobalMaxPool2D())
model_3.add(tf.keras.layers.Dense(50,activation='relu'))
model_3.add(tf.keras.layers.Dense(2,activation='softmax'))
model_3.layers[0].trainable=False#the 0th resnet layer which has already been trained
```

```python
model_3.summary()
```

```python
#Training the model
model_3.compile(loss=tf.keras.losses.BinaryCrossentropy(),optimizer=tf.keras.optimizers.Adam(),metrics=['accuracy'])
```

```python
tf.random.set_seed(100)
import pandas as pd
hist=model_3.fit(train_generator,epochs=2,validation_data=validation_generator)
hist=pd.DataFrame(hist.history).plot();
```
