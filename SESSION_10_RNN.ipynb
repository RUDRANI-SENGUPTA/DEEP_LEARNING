```python
#Stock Market Prediction using LSTM
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
#Accessing the data
adani=pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/ADANIPOWER.NS?period1=1646742140&period2=1678278140&interval=1d&events=history&includeAdjustedClose=true')
```

```python
# use close
data=adani['Close']
```

```python
data
```

```python
#Splitting into train and test
train=data[:200]
test=data[200:]
```

```python
train
```

```python
test
```

```python
#Visualising the data
plt.figure(figsize=(15,6))
plt.plot(train)
plt.plot(test)
plt.xlabel('Market day')
plt.ylabel('Stock price')
plt.legend(['Train','Test']);
```

```python
#Standardisation
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
train_scaled=scaler.fit_transform(np.array(train).reshape(-1,1))
test_scaled=scaler.fit_transform(np.array(test).reshape(-1,1))
```

```python
train_scaled
```

```python
#Converting TS data to regression data
X_train=[]
y_train=[]
#window_size=10
for i in range(10,200):
  X_train.append(train_scaled[i-10:i,0])
  y_train.append(train_scaled[i])
```

```python
X_train
```

```python
y_train
```

```python
X_train.shape
```

```python
#Converting to np.array
X_train=np.array(X_train)
y_train=np.array(y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#Converting TS data to regression data
X_test=[]
y_test=[]
#window_size=10
for i in range(10,50):
  X_test.append(train_scaled[i-10:i,0])
  y_test.append(train_scaled[i])
```

```python
X_test=np.array(X_test)
y_test=np.array(y_test)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
#Model Building
lstm=tf.keras.Sequential()
lstm.add(tf.keras.layers.LSTM(50,return_sequences=True,input_shape=(X_train.shape[1],1)))#(10,1)
lstm.add(tf.keras.layers.Dense(1))
```

```python
#Compiling
lstm.compile(loss=tf.keras.losses.MeanSquaredError(),optimizer=tf.keras.optimizers.Adam(),metrics=['mae'])
```

```python
tf.random.set_seed(10)
hist=lstm.fit(X_train,y_train,epochs=100)
pd.DataFrame(hist.history).plot();
```

```python
#Evaluation of the model
lstm.evaluate(X_test,y_test)
```

```python
#training loss=0.019,testing loss=0.007,mae=0.065,overfitting
```
